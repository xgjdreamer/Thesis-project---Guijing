```python
import numpy as np
import pymannkendall as mk
import pandas as pd
import glob,os
import statsmodels.api as sm
from scipy import stats
from matplotlib import pyplot as plt
```

```python
er_file = glob.glob("S***.csv")
```

```python
dl_er = []
out = []
for f in er_file:
    dl_er.append(pd.read_csv(f))

for x in dl_er:
    print(len(x))
    if len(x)>8:
        x = x.drop(index=[len(x)-1])
        out.append(mk.original_test(x['av_er_ratio']))
```

```python
df = pd.concat(dl_er)
df.to_csv('group3_summary.csv', index = False)
```

```python
res = []
for x in dl_er:
    if len(x)>8:
        x = x.drop(index=[len(x)-1])
        mod = sm.GLM(x['av_er_ratio'], x['end'], family=sm.families.Gaussian(sm.families.links.log))
        res.append(mod.fit())
```

```python
for r in res:
    print(r.summary())
```
